<h2 style="text-align:center;color:green">Add link into main link</h2>

```python
comman_link='https://www.flipkart.com/search?q=best+laptops+under+Rs+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='
for i in range(2,42):
    link=comman_link+str(i)
    page_data=urlopen(link)
    webdata=page_data.read()
    first_page_data+=webdata
```

<h2 style="text-align:center;color:green">Convert into beautifulsoup</h2>

```python
soupdata=soup(first_page_data,'html.parser')  # like str(738437)
type(soupdata)
```

```python
containers=soupdata.findAll('div',{"class":"_2kHMtA"}) # type(container)==It is as list  # containers having all page data which having class is this
print(len(containers),type(containers))
```

<h2 style="text-align:center;color:red">_____Fetch data write into csv file____</h2>

```python
f=open('laptoppage_data.csv','wb')   # writing in binary mode with particular csv file
f.write("ProductName,Stars,Ratings,Reviews,CurrentPrice,MRP,Processor,RAM,Storage,ImageURL\n".encode()) #give a heading
for container in containers:
    # productname
    product=container.find('div',{'class':'_4rR01T'})
    ProductName=product.text.split('-')[0].strip()     # like (str.split(','))   & strip use for remove a space from both side

    # star rating 
    star=container.find('div',{'class':'_3LWZlK'})
    try:
        Stars=star.text   # print only tag text
    except:
        Stars=0

    # rating & reviews
    Rating=container.find("span",{'class':'_2_R_DZ'})
    try:
        ratRev=re.findall('\d+,?\d*',Rating.text)    # \d  for integer ,+ for 1 and more occurence ,? 0 or more occurence,* 0 or more
        Ratings=ratRev[0].replace(",","")
        Reviews=ratRev[1].replace(",","")   # replace "," with " " in text
    except:
        Ratings=0
        Reviews=0
      
    # current price
    CurrentPrice=container.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(",","").replace('₹','')
    
    # mrp
    mrp=container.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP=mrp.text.replace(",","").replace('₹','')
    except:
        MRP=0

    #decription info
    Info=container.findAll('li','rgWa7D')
    if (Info[0].text.endswith(')')|Info[0].text.endswith('Processor')):  # for only useful information
        Processor=Info[0].text
        RAM=Info[1].text
        Storage=Info[3].text
    else:
        Processor=Info[1].text
        RAM=Info[2].text
        Storage=Info[4].text
   
    
    # image url
    Image=container.img
    ImageURL=Image.get('src')
    
    print(ProductName,Stars,Ratings,Reviews,CurrentPrice,MRP,Processor,RAM,Storage,ImageURL)
    f.write(f"{ProductName},{Stars},{Ratings},{Reviews},{CurrentPrice},{MRP},{Processor},{RAM},{Storage},{ImageURL}\n".encode())
    print('\n')
f.close()
```

<h2 style="text-align:center;color:red">____Convert dataframe data into mysql database table____</h2>

```python
def mysqldatabase_connection():
    hostname=input('Enter hostname= ')
    username=input('Enter username= ')
    pwd=input('Enter pwd= ')
    database_name=input('Enter database_name= ')
    charset_value=input('Enter charset_value= ')
    return f"{hostname}",f"{username}",f"{pwd}",f"{database_name}",f"{charset_value}"
hostname,username,pwd,database_name,charset_value=mysqldatabase_connection()

# host='localhost',user='root',password='yoursql password',database='data_scrapper',charset='utf8'
```

```python
### first create database and also table column
import mysql.connector
from mysql.connector import Error

def dataframetosql(dataframe):
    try:
        conn=mysql.connector.connect(host=hostname,user=username,password=pwd,database=database_name,charset=charset_value)   
        mycursor=conn.cursor() 
        # excute query to clear all records from table 
        mycursor.execute("delete from laptopdata")
        conn.commit() #permant saving of data
        for(row,rs) in dataframe.iterrows():
            ProductName=rs[0]   # already in str 
            Stars=str(float(rs[1]))   # for convert str
            Ratings=str(int(rs[2]))
            Reviews=str(int(rs[3]))
            CurrentPrice=str(int(rs[4]))
            MRP=str(int(rs[5]))
            Processor=rs[6]
            RAM=rs[7]
            Storage=rs[8]
            ImageURL=rs[9]
            
            query="insert into laptopdata values(""'"+ProductName+"'"+","+Stars+","+Ratings+","+Reviews+","+CurrentPrice+","+MRP+","+"'"+Processor+"'"+","+"'"+RAM+"'"+","+"'"+Storage+"'"+","+"'"+ImageURL+"'"")"
            mycursor.execute(query)
        conn.commit()
        mycursor.close()
    except Error as e:
        print("Error in mysqlconnection=",e)
    finally:
        if conn.is_connected():
            conn.close()

            
#main program

data_csv=pd.read_csv("laptoppage_data.csv",error_bad_lines=False)
print(data_csv)
dataframetosql(data_csv)
print("\n DataFrame to MYSQL Data Transferd Successfully")
```
